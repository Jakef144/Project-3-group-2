```python
import pandas as pd
import json
```

```python
df = pd.read_csv('qcew-2020-2022.csv')
#print(df.shape)
df.head()
```

```python
annual_df = df[df['Time Period'] == 'Annual']
annual_df
```

```python
first_clean = annual_df.drop_duplicates(subset=['Area Name', 'Year', 'Industry Name'])
cleaned_annual = first_clean.drop_duplicates(subset=['Area Name', 'Year', 'Average Weekly Wages', 'Average Monthly Employment', 'Establishments'])
cleaned_annual
```

```python
cleaned_annual.rename(columns={
    'Area Type':'Area_Type',
    'Area Name':'Area_Name',
    'Time Period':'Time_Period',
    'NAICS Level':'NAICS_Level',
    'NAICS Code':'NAICS_Code',
    'Industry Name':'Industry_Name',
    'Average Monthly Employment':'Average_Monthly_Employment',
    '1st Month Emp':'1st_Month_Emp',
    '2nd Month Emp':'2nd_Month_Emp',
    '3rd Month Emp':'3rd_Month_Emp',
    'Total Wages (All Workers)':'Total_Wages',
    'Average Weekly Wages':'Average_Weekly_Wages'
}, inplace=True)
cleaned_annual
```

```python
top_10_df = cleaned_annual.groupby(['Year', 'Area_Name']).apply(lambda x: x.nlargest(10, 'Average_Weekly_Wages')).reset_index(drop=True)
top_10_df
```

```python
top10_json = top_10_df.to_json(orient='records')
clean_top10_json = json.dumps(json.loads(top10_json), indent=4)
print(clean_top10_json)
```

```python
with open('top10.json', 'w') as json_file:
    json_file.write(clean_top10_json)
```
